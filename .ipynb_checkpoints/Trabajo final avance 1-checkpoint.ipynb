```python
# Esta funcion obtiene la lista de nombres, cantidad y precio unitario de plaza vea

def obtener_info_plazavea(driver):

    try:
        lista1 = driver.find_elements(by=By.XPATH, value='//div[@class="HA Showcase Showcase--food"]')
    except NoSuchElementException:
        lista1 =[]

    try:
        lista2 = driver.find_elements(by=By.XPATH, value='//div[@class="HA Showcase Showcase--non-food"]')
    except NoSuchElementException:
        lista2 =[]   
        
    lista = lista1 + lista2
    
    print(f'Se encontraron {len(lista)} productos relacionados con el nombre.')
    print('Iniciando scraping...')
    
    info = []
    df = pd.DataFrame()
    for producto in lista:
        
        try:
            nombre =  producto.find_element(by=By.XPATH, value='.//a[@class="Showcase__name"]').text
        except NoSuchElementException:
            nombre ="NN"
        
        try:            
            precio = producto.find_element(by=By.XPATH, value='.//span[@class="price"]').text
        except NoSuchElementException:
            precio ="NN"
            
        try:
            cantidad =  producto.find_element(by=By.XPATH, value='.//div[@class="Showcase__units-reference"]').text
        except NoSuchElementException:
            cantidad ="NN"          
            
        info.append({'nombre':nombre, 'cantidad': cantidad,  'precio': precio})
        
        df = pd.DataFrame(info)
    return df
```

```python
obtener_info_plazavea(driver)
```
