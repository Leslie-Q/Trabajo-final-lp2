```python
# En este bloque importaremos todas las librerias a usar
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# Inspeccionando la página de plaza vea con el producto azúcar
url ='https://www.plazavea.com.pe/search/?_query=azucar'
r = requests.get(url)
r.status_code
s = BeautifulSoup(r.text, 'lxml')
print(s.prettify())
```

```python
# Instanciando el driver del navegador chrome
PATH = 'C:/Users/Leslie/Desktop/cuaderno/TRABAJO FINAL LP2 Leslie/chromedriver.exe'
options = webdriver.ChromeOptions()
options.add_argument('--incognito')
driver = webdriver.Chrome(PATH,options = options)
```

```python
# Direccionando a la página de búsqueda
driver = webdriver.Chrome(PATH,options = options)
driver.get(url)
```

```python
# Almacenando la lista con la descripción común encontrada en plaza vea
lista = driver.find_elements(by=By.XPATH, value='//div[@class="HA Showcase Showcase--food"]')
lista
```

```python
# Almacenando el elemento 3 de la lista
producto = lista[2]
producto
```

```python
# Encontrando  el precio 
precio  = producto.find_element(by=By.XPATH, value='.//div[@data-price]')
precio.text
```

```python
# Encontrando el nombre 
nombre  = producto.find_element(by=By.XPATH, value='.//div[@title]')
nombre.text
```
